```python
import pandas as pd

game=pd.read_csv(r"C:\Users\Bryant Xia\Desktop\lead_change_analysis\playoffs_22\g2022054_PHI_MIA.csv")

game.head(50)
```

```python
import matplotlib.pyplot as plt 
```

```python
prev="0-0"
score_trimmed=[]
away=[]
home=[]
for elt in game['score']:
    if (elt!=prev):
        prev=elt
        split=elt.split("-")
        away.append(int(split[0]))
        home.append(int(split[1]))
        score_trimmed.append(elt)
        
print(len(game['score'])-len(score_trimmed))
```

```python
plt.plot(home)
plt.plot(away)
```
