```python
import pandas as pd
april=pd.read_csv(r"C:\Users\Bryant Xia\Desktop\lead_change_analysis\may_games.csv")
april.tail(10)
```

```python
ct=0
for dt in april['Date']:
    april['Date'][ct]=dt[8:len(dt)]
    split_string = april['Date'][ct].split(",", 1)
    april['Date'][ct]=int(split_string[0])
    ct+=1
```

```python
april.shape
```

```python
abbrev=pd.read_csv(r"C:\Users\Bryant Xia\Desktop\lead_change_analysis\team_abbrev.csv")
dict={}
for i in range (0, 30):
    dict[abbrev['Team'][i]]=abbrev['Abbrev'][i]
april['home_abb']=""
april['away_abb']=""
for i in range (0,len(april)):
    april['home_abb'][i]=dict[april['Home/Neutral'][i]]
    april['away_abb'][i]=dict[april['Visitor/Neutral'][i]]
```

```python
april.iloc[24]
```

```python
april['url']=""
for i in range (0, len(april)):
    if (april['Date'][i]>9):
        april['url'][i]="https://www.basketball-reference.com/boxscores/shot-chart/202205"+str(april['Date'][i])+"0"+april['home_abb'][i]+".html"
    else:
        april['url'][i]="https://www.basketball-reference.com/boxscores/shot-chart/2022050"+str(april['Date'][i])+"0"+april['home_abb'][i]+".html"
```
