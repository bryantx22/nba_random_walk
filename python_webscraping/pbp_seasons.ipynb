#### try working on just one url to make sure we have the right code

```python
def scrape (dest, id):

    url=dest

    content=requests.get(url).text
    soup = BeautifulSoup(content, "lxml")

    pbp_table = soup.find("table", attrs={"id": "pbp"})
    pbp_rows=pbp_table.find_all("tr")

    time=[]
    home_des=[] #description
    home_change=[]
    score=[]
    away_change=[]
    away_des=[]

    for i in range (0,len(pbp_rows)):
        row=pbp_rows[i].find_all('td')
        if (len(row)!=6):
            pass
        else:
            time.append(row[0].text)
            away_des.append(row[1].text)
            away_change.append(row[2].text)
            score.append(row[3].text)
            home_change.append(row[4].text)
            home_des.append(row[5].text)

    time=pd.Series(time)
    away_des=pd.Series(away_des)
    away_change=pd.Series(away_change)
    score=pd.Series(score)
    home_change=pd.Series(home_change)
    home_des=pd.Series(home_des)

    data=[time, away_des, away_change, score, home_change, home_des]
    col_names=["time","away_description","away_change","score","home_change","home_description"]
    df = pd.concat(data, axis=1, keys=col_names)

    file_name=r"C:\Users\Bryant Xia\Desktop\lead_change_analysis\pbp_games\g"+id+".csv"
    df.to_csv(file_name, index=False)
```

```python
scrape (games['urls'][0], games['id'][0])
```
